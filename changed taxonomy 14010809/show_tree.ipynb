```python
from treelib import Node,Tree
```

```python
import pandas as pd
```

```python
df=pd.read_excel('710FN1-DEHESTANI1.xlsx')
```

```python
for index , row in df.iterrows():
    print(row[2]) #name joze
    print(row[6]) #serial nomber 
    print(row[7]) #sreial balasari
    
    
```

```python
tree = Tree()
```

```python
i=0
for index , row in df.iterrows():
    print(i)
    if i == 0:
        tree.create_node(row[2],row[6])
    else:
        tree.create_node(row[2],row[6],parent=row[7])
    i += 1
    
```

```python
tree.show()
```

```python
fname=format(df.iloc[0,2]+'.txt')
print(fname)
tree.save2file(fname)
```
